### Les flottants (ou nombres  virgule flottante)

Les nombres  virgule flottante sont des nombres comportant une partie
entire et une partie dcimale aprs la virgule marque par un point
(notation anglo-saxonne).

Entrez

``` python
351.4 ** 6
```

Lorsque la partie entire dun nombre flottant devient trop grande ou
que le nombre commence par une grande suite de zros (y compris la
partie entire) : le nombre est donne en notation scientifique (comme
sur vos calculatrices).  
Ici **e** signifie *fois dix puissance* : $\times10^{n}$.

``` python
15.1 ** -8
```

On peut dailleurs directement crire les nombres flottants sous cette
forme :

``` python
6.02e21
```

*Valeur exacte ou approximation ?*

Entrez :

``` python
0.20000000000000000000000458
0.00000000002458652541255612
0.000000000024586525412556124654564564564564545645
```

Attention donc : la valeur retourne (et donc ventuellement stocke)
par Python peut tre une approximation du rel entr.

Enfin, il est intressant de noter que Python convertit implicitement
les entiers en flottants si ncessaire :

``` python
4 + 5.2
```

**Remarque :** chaque oprateur existe sous forme dune fonction
(cf.documentation officielle : Correspondance entre oprateurs et
fonctions
<a href="https://docs.python.org/fr/3.7/library/operator.html?highlight=op%C3%A9rateurs%20affectation#mapping-operators-to-functions" class="uri">https://docs.python.org/fr/3.7/library/operator.html?highlight=op%C3%A9rateurs%20affectation#mapping-operators-to-functions</a>)

\_\_ mettre ici paragraphe sur valeur max et min possible avant infini

### Chaines de caractres et oprations

Testez dans le shell les oprations suivantes :

``` python
'abc' + 'def'
'abc' * 2
'abc' **3
'abc' / 2
'abc' // 2 
```
